

\#1 What is hypothesis testing in statistics?
- Hypothesis testing is a statistical method used to make decisions or inferences about population parameters.
It involves testing an assumption (hypothesis) using sample data.
The outcome determines whether to reject or not reject the null hypothesis.

---

\#2 What is the null hypothesis, and how does it differ from the alternative hypothesis?
- The null hypothesis (H₀) states there is no effect or difference.
The alternative hypothesis (H₁) suggests there is an effect or difference.
Hypothesis testing assesses whether data provides enough evidence to reject H₀.

---

\#3 What is the significance level in hypothesis testing, and why is it important?
- The significance level (α) is the threshold for rejecting the null hypothesis, commonly set at 0.05.
It represents the probability of making a Type I error.
Lower α reduces the chance of false positives but may increase false negatives.

---

\#4 What does a P-value represent in hypothesis testing?
- A P-value indicates the probability of obtaining results as extreme as the observed ones, assuming H₀ is true.
It measures the strength of evidence against the null hypothesis.
Smaller P-values suggest stronger evidence to reject H₀.

---

\#5 How do you interpret the P-value in hypothesis testing?
- If the P-value is less than the significance level (α), reject the null hypothesis.
If it is greater, fail to reject H₀.
It helps quantify the probability of observing the data under the null assumption.

---

\#6 What are Type 1 and Type 2 errors in hypothesis testing?
- Type 1 error occurs when H₀ is wrongly rejected (false positive).
Type 2 error happens when H₀ is wrongly accepted (false negative).
Balancing these errors is key to reliable hypothesis testing.

---

\#7 What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
- A one-tailed test checks for an effect in only one direction (greater or less).
A two-tailed test checks for an effect in both directions.
Choice depends on the research question.

---

\#8 What is the Z-test, and when is it used in hypothesis testing?
- A Z-test is used to compare sample and population means when population standard deviation is known.
It’s suitable for large samples (n > 30) and normally distributed data.
It uses the Z-score for test statistics.

---

\#9 How do you calculate the Z-score, and what does it represent in hypothesis testing?
- Z = (X̄ - μ) / (σ / √n), where X̄ is the sample mean, μ is the population mean, and σ is the standard deviation.
It shows how many standard deviations a value is from the mean.
It helps determine the probability of the result under H₀.

---

\#10 What is the T-distribution, and when should it be used instead of the normal distribution?
- The T-distribution is used when the sample size is small (n < 30) and population standard deviation is unknown.
It is similar to the normal distribution but has heavier tails.
It adjusts for extra variability in small samples.

---

\#11 What is the difference between a Z-test and a T-test?
- A Z-test is used with large samples and known population standard deviation.
A T-test is used with small samples and unknown standard deviation.
The T-test accounts for more variability due to smaller samples.

---

\#12 What is the T-test, and how is it used in hypothesis testing?
- A T-test compares sample means to test for significant differences.
It’s used when the population standard deviation is unknown.
There are one-sample, two-sample, and paired T-tests.

---

\#13 What is the relationship between Z-test and T-test in hypothesis testing?
- Both are used to test hypotheses about means.
Z-tests assume known standard deviation and large samples, while T-tests do not.
As sample size increases, the T-distribution approaches the Z-distribution.

---

\#14 What is a confidence interval, and how is it used to interpret statistical results?
- A confidence interval provides a range of plausible values for a population parameter.
It is associated with a confidence level (e.g., 95%).
It helps assess the precision and reliability of an estimate.

---

\#15 What is the margin of error, and how does it affect the confidence interval?
- Margin of error indicates the range above and below the sample estimate in a confidence interval.
A larger margin means less precision.
It’s affected by sample size and variability.

---

\#16 How is Bayes' Theorem used in statistics, and what is its significance?
- Bayes' Theorem updates the probability of a hypothesis based on new evidence.
It is expressed as P(A|B) = \[P(B|A) × P(A)] / P(B).
It’s widely used in decision-making, medical testing, and machine learning.

---

\#17 What is the Chi-square distribution, and when is it used?
- The Chi-square distribution is used for categorical data analysis.
It’s applied in goodness-of-fit tests and tests for independence.
It’s positively skewed and defined by degrees of freedom.

---

\#18 What is the Chi-square goodness of fit test, and how is it applied?
- This test checks if observed frequencies match expected frequencies in categorical data.
It determines whether a sample fits a particular distribution.
A high Chi-square statistic indicates a poor fit.

---

\#19 What is the F-distribution, and when is it used in hypothesis testing?
- The F-distribution is used to compare variances of two or more groups.
It is right-skewed and defined by two sets of degrees of freedom.
Used mainly in ANOVA and variance ratio tests.

---

\#20 What is an ANOVA test, and what are its assumptions?
- ANOVA (Analysis of Variance) compares means across three or more groups.
Assumptions: normality, independence, and equal variances.
It tests whether at least one group mean is different.

---

\#21 What are the different types of ANOVA tests?
- There are one-way, two-way, and repeated measures ANOVA.
One-way tests one factor, two-way tests interaction between two factors.
Repeated measures test the same group over time or conditions.

---

\#22 What is the F-test, and how does it relate to hypothesis testing?
- The F-test compares two population variances to see if they are significantly different.
It’s used in ANOVA to assess overall group differences.
A significant F-test suggests at least one group differs in mean.

```python
# 1 Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results
import numpy as np
from scipy.stats import norm

sample = np.random.normal(loc=102, scale=10, size=50)
sample_mean = np.mean(sample)
pop_mean = 100
pop_std = 10

z = (sample_mean - pop_mean) / (pop_std / np.sqrt(len(sample)))
p_value = 1 - norm.cdf(z)

print(f"Z-score: {z:.2f}")
print(f"P-value: {p_value:.4f}")
```

```python
# 2 Simulate random data to perform hypothesis testing and calculate the corresponding P-value using Python
from scipy import stats

data = np.random.normal(55, 5, 100)
t_stat, p_value = stats.ttest_1samp(data, popmean=50)
print(f"T-statistic: {t_stat:.2f}, P-value: {p_value:.4f}")
```

```python
# 3 Implement a one-sample Z-test using Python to compare the sample mean with the population mean
def z_test(sample, pop_mean, pop_std):
    sample_mean = np.mean(sample)
    z = (sample_mean - pop_mean) / (pop_std / np.sqrt(len(sample)))
    return z, norm.sf(abs(z)) * 2

sample_data = np.random.normal(52, 10, 40)
z, p = z_test(sample_data, 50, 10)
print(f"Z = {z:.2f}, P-value = {p:.4f}")
```

```python
# 4 Perform a two-tailed Z-test using Python and visualize the decision region on a plot
import matplotlib.pyplot as plt

z_critical = norm.ppf(0.975)  # for 95% confidence
x = np.linspace(-4, 4, 1000)
plt.plot(x, norm.pdf(x), label='Standard Normal')

plt.axvline(z_critical, color='red', linestyle='--', label='Critical Value')
plt.axvline(-z_critical, color='red', linestyle='--')
plt.fill_between(x, 0, norm.pdf(x), where=(x > z_critical) | (x < -z_critical), color='red', alpha=0.3)

plt.title("Two-Tailed Z-test")
plt.legend()
plt.show()
```

```python
# 5 Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing
def type_errors(mu0, mu1, sigma, n, alpha=0.05):
    z_critical = norm.ppf(1 - alpha)
    x = np.linspace(mu0 - 4*sigma, mu1 + 4*sigma, 1000)
    plt.plot(x, norm.pdf(x, mu0, sigma/np.sqrt(n)), label="H0 Distribution")
    plt.plot(x, norm.pdf(x, mu1, sigma/np.sqrt(n)), label="H1 Distribution")
    critical_value = mu0 + z_critical * (sigma / np.sqrt(n))
    plt.axvline(critical_value, color='red', linestyle='--', label="Critical Value")
    plt.fill_between(x, 0, norm.pdf(x, mu0, sigma/np.sqrt(n)), where=(x > critical_value), color='red', alpha=0.3, label="Type I Error")
    plt.fill_between(x, 0, norm.pdf(x, mu1, sigma/np.sqrt(n)), where=(x <= critical_value), color='blue', alpha=0.3, label="Type II Error")
    plt.legend()
    plt.title("Type I and II Errors")
    plt.show()

type_errors(100, 105, 15, 30)
```

```python
# 9 Write a Python function to calculate the confidence interval for a sample mean and explain its significance
def confidence_interval(data, confidence=0.95):
    mean = np.mean(data)
    se = stats.sem(data)
    interval = stats.t.interval(confidence, len(data)-1, loc=mean, scale=se)
    return interval

sample = np.random.normal(50, 5, 30)
print("Confidence Interval:", confidence_interval(sample))
```

```python
# 10 Write a Python program to calculate the margin of error for a given confidence level using sample data
data = np.random.normal(100, 15, 50)
confidence = 0.95
se = stats.sem(data)
moe = se * stats.t.ppf((1 + confidence) / 2., df=len(data)-1)
print("Margin of Error:", moe)
```

```python
# 11 Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process
def bayes_theorem(prior_A, prior_B, likelihood_A, likelihood_B):
    numerator = likelihood_A * prior_A
    denominator = likelihood_A * prior_A + likelihood_B * prior_B
    return numerator / denominator

posterior = bayes_theorem(0.01, 0.99, 0.95, 0.05)
print("Posterior Probability of A:", posterior)
```
